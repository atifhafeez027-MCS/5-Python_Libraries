```python
import numpy  as np
import pandas as pd
```

```python
diabetes_data = pd.read_csv('diabetes.csv')
```

### Pandas Questions

```python
#Q1. Load the dataset and display the first 10 rows.
DF = diabetes_data
DF.head(10)
```

```python
#Q2. Show the summary statistics for all numerical columns.

DF.describe()
```

```python
#Q3 How many missing (zero) values are in the 'Insulin' column?

DF[DF['Insulin'] == 0]
(DF['Insulin'] == 0).sum()
#There are 374 columns where value of insulin is zero
```

```python
#Q4 Create a new column 'BMI_Class' categorizing BMI values into 'Underweight', 'Normal', 'Overweight', and 'Obese'.

def bmi_cat(BMI):
    if BMI < 18.5:
        return "Underweight"
    elif BMI <  25:
        return "Normal"
    elif BMI < 30:
        return "Overweight"
    else:
        return "Obese"
        
DF["BMI_Class"] =  DF["BMI"].apply(bmi_cat)
DF
```

```python
#Q5 Group the dataset by 'Outcome' and compute the average age for each group.

group_by_outcome = DF.groupby('Outcome')['Age'].mean()
print(group_by_outcome)
```

```python
#Q6 Filter the data to include only patients with a 'Glucose' level above 120.

Gluc_above_120 = DF[DF['Glucose'] > 120]
Gluc_above_120
```

```python
#Q7 Using the `query` method, select rows where Age is greater than 30 and BMI is less than 30.

DF.query('Age > 30 and BMI < 30')
```

```python
#Q8 Drop rows where any of the values in ['Glucose', 'BloodPressure', 'BMI'] are zeros.

DF.drop(DF[(DF['Glucose'] == 0) | (DF['BloodPressure'] == 0) | (DF['BMI'] == 0)].index, inplace=True)
DF
#.index is used to access the index labels of the rows that meet the condition specified within the parentheses.
```

```python
#Q9 Reset the index of the dataframe after dropping rows.

DF.reset_index(drop=True, inplace=True)
DF
```

```python
#Q10  Save a subset of the dataset with columns ['Age', 'Glucose', 'Outcome'] to a new CSV file.

Subset_DF = DF[['Age', 'Glucose', 'Outcome']]
Subset_DF
# Saving the subset df to a new CSV file
Subset_DF.to_csv('diabetes_subset_data.csv', index=False)
```

### NumPy Questions

```python
import numpy as np
```

```python
#Q1 How many unique values are there in the 'Outcome' column using NumPy?

unique_in_outcome = np.unique(DF['Outcome'].values)
unique_in_outcome = len(np.unique(DF['Outcome'].values))

print("Number of Unique values in the Outcome column are:", unique_in_outcome)
```

```python
#Q2 Compute the mean and standard deviation of 'Glucose' levels using NumPy.

glucose_mean = np.mean(DF['Glucose'])
print(glucose_mean)

glucose_std = np.std(DF['Glucose'])
print (glucose_std)
```

```python
#Q3 Create a NumPy array from the 'BMI' column and find the median.

BMI = DF['BMI'].to_numpy() 
BMI
BMI_Median = np.median(BMI)
BMI_Median
```

```python
#Q4 Using NumPy, count how many patients are there for each unique number of pregnancies.

unique_preg, counts = np.unique(DF['Pregnancies'], return_counts=True)

# We will display the counts using for loop
for pregnancies, count in zip(unique_preg, counts):
    print(f"Number of patients with {pregnancies} pregnancies: {count}")
```

```python
#Q5 Determine the maximum and minimum insulin level using NumPy.

Max_insulin = np.max(DF['Insulin'])
print("Max insulin leve: ",Max_insulin)

Min_insulin = np.min(DF['Insulin'])
print("Min insulin level: ",Min_insulin)
```

```python
#Q6 Calculate the variance of 'Age' in the dataset using NumPy.

Var_age = np.var(DF['Age'])
print("Variance of age in the data set is: ", Var_age)
```

```python
#Q7 Filter the array of 'Glucose' to include only values greater than the mean glucose level.

gluc_val = DF['Glucose'].to_numpy()
gluc_val

gluc_mean = np.mean(DF['Glucose'])
gluc_mean

req_val = gluc_val[gluc_val > gluc_mean]
req_val
```

```python
#Q8 Using NumPy, create a function to normalize the 'BloodPressure' values (subtract the mean and divide by the standard deviation).

def BloodPressure(BP_values):
    mean_bp = np.mean(BP_values)
    std_bp = np.std(BP_values)
    Normal_BP = (BP_values - mean_bp) / std_bp
    return Normal_BP

BP_values = DF['BloodPressure'].to_numpy()
BP = BloodPressure(BP_values)
print(BP)
```

```python
#Q9 Compute the 25th percentile of 'DiabetesPedigreeFunction' using NumPy.

Diabetes_Pedigree_Function = np.percentile(DF['DiabetesPedigreeFunction'], 25)
Diabetes_Pedigree_Function
```

### MATPLOTLIB Questions

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Q1 Plot a histogram of 'Age'.

plt.hist(DF['Age'], bins = 8, color = 'blue', edgecolor= 'red')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show() 
```

```python
#Q2 Create a scatter plot showing 'Glucose' versus 'Insulin' levels.

plt.scatter(DF['Glucose'], DF['Insulin'], color = 'blue', alpha = 0.5)

plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.title('Scatter Plot of Glucose vs Insulin')

plt.show()
```

```python
#Q3 Draw a bar chart comparing the mean 'BMI' across 'Outcome'.

mean_BMI = DF.groupby('Outcome')['BMI'].mean()

plt.bar(mean_BMI.index, mean_BMI, color = ['red', 'green'])

plt.xlabel('Outcome')
plt.ylabel('Mean BMI')
plt.title('Mean BMI Across Outcome')

plt.legend(['Not Diabetes', 'Diabetes'])
plt.show()
```

```python
#Q4 Plot a line graph of 'BloodPressure' sorted by 'Age'.

Age_BP = DF.sort_values(by = 'Age')

plt.plot(Age_BP['Age'], Age_BP['BloodPressure'], color = 'red')

plt.xlabel('Age')
plt.ylabel('BloodPressure')
plt.title('Line Graph of BloodPressure Sorted by Age')

plt.show()
```

```python
#Q5 Display a box plot of 'BMI' grouped by 'Outcome'.

DF.boxplot(column = "BMI", by = "Outcome")
plt.show()
```

```python
#Q6 Create a pie chart showing the proportion of diabetes outcomes (0's and 1's).

diabetes_outcome = DF['Outcome'].value_counts()

plt.pie(diabetes_outcome, labels=diabetes_outcome.index, autopct='%1.1f%%', colors=['darkred', 'green'])

plt.title('Diabetes Outcomes')

plt.show()
```

```python
#Q7 Plot a stacked bar chart of 'Pregnancies' categorized by 'Outcome'.

preg_outcome = DF.groupby(['Outcome', 'Pregnancies']).size().unstack()

preg_outcome.plot(kind = 'bar', stacked = True, color = ['red', 'green'])

plt.xlabel('Pregnancies')
plt.ylabel('Count')
plt.title('Stacked Bar Chart of Pregnancies Categorized by Outcomes')

plt.show()
```

```python
#Q8 Create subplots showing histograms for 'Glucose' and 'BloodPressure'.

fig, axs = plt.subplots(1, 2, figsize = (9, 4))

# Histogram for Glucose
axs[0].hist(DF['Glucose'], bins = 10, color = 'red', edgecolor = 'blue')
axs[0].set_title('Glucose Histogram')
axs[0].set_xlabel('Glucose')
axs[0].set_ylabel('Frequency')

# Histogram for Blood Pressure
axs[1].hist(DF['BloodPressure'], bins = 10, color = 'red', edgecolor = 'blue')
axs[1].set_title('Blood Pressure Histogram')
axs[1].set_xlabel('Blood Pressure')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Q9 Generate a heat map of the correlation matrix of the dataset.
import seaborn as sb

corel_matrix = DF.corr(numeric_only = True)

plt.figure(figsize=(10, 8))
sb.heatmap(corel_matrix, annot = True, cmap = 'coolwarm', fmt = ".2f", linewidths = 0.5)
plt.title('Co-Relation Matrix Heatmap')
plt.show()
```
